Q1 TOP 5 COUNTRIES WITH THE MOST SALES

```python
top_5 = df_project.groupby(['country']).country_id.count().sort_values(ascending =False)[:5]
top_5
```

```python
#Top 5 countries according to production
plt.figure(figsize =(12,8))
sns.barplot(x = top_5.index,y=top_5.values)
plt.xlabel('Country')
plt.ylabel('Numbers of records in the dataset')
plt.title('Top 5 countries by most data records');
```

Q2 TOP 10 MOST REPORTED COMMODITIES

```python
com_10 = df_project.groupby(['produce']).produce_id.count().sort_values(ascending =False)[:10]
com_10
```

```python
ax = com_10.plot(kind='bar', y= 'com_10.index',
                legend= False,figsize=(12,8), fontsize=10)
ax.set(title = 'Top 10 most reported commodities',
      xlabel = 'produce',
      ylabel = 'rowcount');
#ax.set_xlabel("produce", fontsize=10)
#ax.set_ylabel("rowcount", fontsize=10)
plt.style.use('ggplot');
```

Q4. DIFFERENT KIND OF PRODUCTS PRODUCED FROM 1990 - 2021

```python
duplicate = df_project.drop_duplicates(['country','produce'], keep = 'last') # dropping duplicate values
produce_count = duplicate.groupby(['country']).count()
p = produce_count.sort_values('produce_id', ascending = False)[:5]
p['produce_id']
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,10))
g = sns.barplot(x=p.produce_id,y=p.index,data=p)
plt.title('Produce count for each country')
plt.xlabel('Produce_count')
plt.ylabel('Country')
plt.show()
```

```python
Ethiopia has the highest numbers of different product produced over the years from 1990 - 2021
```

Q5 PRODUCT PRODUCED PER YEAR IN AFRICA

```python
duplicate1 = df_project.drop_duplicates(['year','produce'],keep = 'last') # dropping duplicate values
produce_count = duplicate1.groupby(['year']).count()
a = produce_count.sort_values('produce_id', ascending =False)
a['produce_id'].head()
```

```python
# visualize the amount of different kind of product produced per year in africa
plt.figure(figsize=(20,8))
sns.barplot(x = a.index, y= a.produce_id,data = a)
plt.xlabel('Year',fontsize = 15)
plt.ylabel('Count',fontsize = 15)
plt.title('Count of different product produced per year in AFRICA');
```
